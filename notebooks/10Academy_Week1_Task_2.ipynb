```python
import pandas as pd
import yfinance as yf
import numpy as np
import os
```

```python
# Path to raw data
raw_data_path = "../data/raw_analyst_ratings.csv"

# Try to read the raw data
try:
    raw_data = pd.read_csv(raw_data_path)
except FileNotFoundError:
    print(f"Error: File not found at {raw_data_path}.")
    raw_data = None
```

```python
# If data is loaded successfully
if raw_data is not None:
    # Convert 'date' to datetime and handle missing or infinite values
    raw_data['date'] = pd.to_datetime(raw_data['date'], errors='coerce')
    raw_data['date'] = raw_data['date'].replace([np.inf, -np.inf], np.nan)
    raw_data['date'].ffill(inplace=True)  # Forward-fill missing dates

    # Check for missing values in 'date'
    missing_dates = raw_data['date'].isnull().sum()
    if missing_dates > 0:
        print(f"Warning: {missing_dates} missing dates after forward-filling.")
    else:
        print("No missing dates after forward-filling.")

    # Extract unique stock symbols and determine date range
    unique_stocks = raw_data['stock'].unique()
    earliest_date = raw_data['date'].min()
    latest_date = raw_data['date'].max()

    print(f"Unique stock symbols: {unique_stocks}")
    print(f"Date range for stock data: {earliest_date} to {latest_date}")

    # Create a dictionary to store stock data
    stock_data = {}

    # Download stock prices for each unique stock symbol
    for stock in unique_stocks:
        try:
            # Fetch stock data
            stock_df = yf.download(stock, start=earliest_date, end=latest_date)

            # Check for expected columns
            expected_columns = ['Open', 'Close', 'High', 'Low', 'Volume']
            for col in expected_columns:
                assert col in stock_df.columns, f"Missing '{col}' in stock data."

            # Store in dictionary
            stock_data[stock] = stock_df
            print(f"Downloaded data for {stock}.")

        except Exception as e:
            print(f"Error fetching data for {stock}: {e}")

    output_dir = "../data/stock_prices/"
    os.makedirs(output_dir, exist_ok=True)

    # Save stock data to CSV for later use
    for stock_symbol, stock_df in stock_data.items():
        file_path = os.path.join(output_dir, f"{stock_symbol}_data.csv")
        stock_df.to_csv(file_path, index=True)  # Save with index to keep dates
        print(f"Saved data for {stock_symbol} to {file_path}")
    # Sample data for one of the stock symbols
    if stock_data:
        sample_stock = list(stock_data.keys())[0]
        print(f"Sample data for {sample_stock}:")
        print(stock_data[sample_stock].head())
    else:
        print("No stock data fetched.")

else:
    print("Data loading failed. Unable to proceed with stock price fetching.")
```
